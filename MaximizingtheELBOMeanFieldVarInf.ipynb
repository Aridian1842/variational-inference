#### Maximizing the ELBO ####

Finally, we have our objective function:
    
$$\mathcal{L}(Q) = E_Q \left[{\rm{ln\ }} P(X, Z)\right] - E_Q \left[{\rm{ln\ }} {\prod_{i=1}^{n} Q_i(Z_i)}\right]$$

We want to maximize $\mathcal{L}(Q)$ with respect to each function $Q_i(Z_i)$

First, we can simplify this form a bit:
    
\begin{align}
\mathcal{L}(Q) &= E_Q \left[{\rm{ln\ }} P(X, Z)\right] - E_Q \left[{\rm{ln\ }} {\prod_{i=1}^{n} Q_i(Z_i)}\right] \\[0.5em]
&= E_Q\ {\rm{ln}} \left[\frac{P(X, Z)}{\prod_{i=1}^{n} Q_i(Z_i)} \right] \\[0.5em]
&= \int_{Z_1} \int_{Z_2} \cdots \int_{Z_n} \prod_{i=1}^{n} Q_i(Z_i) {\rm{\ ln}}\left[\frac{P(X, Z)}{\prod_{i=1}^{n} Q_i(Z_i)} \right] dZ_1 dZ_2 \cdots dZ_n\\[0.5em]
&= \int_{Z_1} \int_{Z_2} \cdots \int_{Z_n} \prod_{i=1}^{n} Q_i(Z_i) \left[ {\rm{\ ln}}P(X, Z) - {\rm{\ ln}}\prod_{i=1}^{n} Q_i(Z_i) \right] dZ_1 dZ_2 \cdots dZ_n\\[0.5em]
\end{align}

But remember, each $Z_i$ is independent under the mean field assumption. So we can separate out these products:
    
    
\begin{align}
\int_{Z_1} \int_{Z_2} \cdots \int_{Z_n} Q_j(Z_j) \prod_{i \ne j} Q_i(Z_i) \left[ {\rm{\ ln}}P(X, Z) - {\rm{\ ln}} \left(Q_j(Z_j) \prod_{i \ne j} Q_i(Z_i)\right) \right] dZ_1 dZ_2 \cdots dZ_n\\[0.5em]
\int_{Z_1} \int_{Z_2} \cdots \int_{Z_n} Q_j(Z_j) \prod_{i \ne j} Q_i(Z_i) \left[ {\rm{\ ln}}P(X, Z) - {\rm{\ ln}} Q_j(Z_j) - \sum_{i \ne j} {\rm{\ ln}} Q_i(Z_i) \right]dZ_1 dZ_2 \cdots dZ_n\\[0.5em]
\end{align}

$$\left(\int_{Z_1} \cdots \int_{Z_n} Q_j(Z_j) \prod_{i \setminus j} Q_i(Z_i) {\rm{\ ln}}P(X, Z) dZ_1 \cdots dZ_n \right) - \left( \int_{Z_1} \cdots \int_{Z_n} Q_j(Z_j) \prod_{i \ne j} Q_i(Z_i) {\rm{\ ln}} Q_j(Z_j) dZ_1 \cdots dZ_n \right) - \left(\int_{Z_1} \cdots \int_{Z_n} Q_j(Z_j) \prod_{i \ne j} Q_i(Z_i) \sum_{i \ne j} {\rm{\ ln}} Q_i(Z_i)dZ_1\cdots dZ_n \right)$$

Also because of the independence assuption, and assuming regularity, the $Z_j$ element can be factored out of the integrals:


\begin{align}
\int_{Z_j} Q_j(Z_j) \left\{ \int_{Z_{i \ne j}} \prod_{i \ne j} Q_i(Z_i) {\rm{\ ln}}P(X, Z) dZ_{i \ne j} \right\} dZ_j - \int_{Z_j} Q_j(Z_j) {\rm{\ ln}} Q_j(Z_j)\left\{ \int_{Z_{i \ne j}} \prod_{i \ne j} Q_i(Z_i)  dZ_{i \ne j} \right\} dZ_j - \int_{Z_j} Q_j(Z_j) \left\{ \prod_{i \ne j} Q_i(Z_i) \sum_{i \ne j} {\rm{\ ln}} Q_i(Z_i) dZ_{i \ne j} \right\} d{Z_j}
\end{align}

The first term can be seen as an expectation with respect to the product of all $Q$ distributions except $Q_j$, in a similar manner to our treatment of the ELBO.

In the second term, recall that each distribution $Q$ is a probability density function (pdf). In fact, we assumed that $Q$ factorized into $n$ independent distributions. So:

$$Q(Z) = \prod_{i=1}^{n} Q_i(Z_i)$$

is a pdf, which means also that:

$$Q(Z_{i \ne j}) = \prod_{i \ne j} Q_i(Z_i)$$

is a pdf. And we know that the integral of a pdf over its domain is 1. So in the second term:

$$\left\{ \int_{Z_{i \ne j}} \prod_{i \ne j} Q_i(Z_i)  dZ_{i \ne j} \right\} = 1$$

In the third term, the term inside the curly braces does not depend on $Z_j$, so it is constant with respect to $Z_j$. Call this term "a" and we see that this term evaluates to a constant:


$$\int_{Z_j} Q_j(Z_j) \left\{ \prod_{i \ne j} Q_i(Z_i) \sum_{i \ne j} {\rm{\ ln}} Q_i(Z_i) dZ_{i \ne j} \right\} d{Z_j}$$

$$\int_{Z_j} Q_j(Z_j) \left\{a \right\} d{Z_j}$$

$$a \int_{Z_j} Q_j(Z_j) d{Z_j}$$

$$a$$

So we come from:

\begin{align}
\mathcal{L}(Q) = \int_{Z_j} Q_j(Z_j) \left\{ \int_{Z_{i \ne j}} \prod_{i \ne j} Q_i(Z_i) {\rm{\ ln}}P(X, Z) dZ_{i \ne j} \right\} dZ_j - \int_{Z_j} Q_j(Z_j) {\rm{\ ln}} Q_j(Z_j)\left\{ \int_{Z_{i \ne j}} \prod_{i \ne j} Q_i(Z_i)  dZ_{i \ne j} \right\} dZ_j - \int_{Z_j} Q_j(Z_j) \left\{ \prod_{i \ne j} Q_i(Z_i) \sum_{i \ne j} {\rm{\ ln}} Q_i(Z_i) dZ_{i \ne j} \right\} d{Z_j}
\end{align}

to:
\begin{align}
\mathcal{L}(Q) = \int_{Z_j} Q_j(Z_j) E_{i \ne j} \left[ {\rm{\ln\ }} P(X, Z) \right] dZ_j - \int_{Z_j} Q_j(Z_j) {\rm{\ ln}} Q_j(Z_j) dZ_j - a
\end{align}

A function of just one of our hidden variables.

Now let ${\rm {ln\ }} \tilde{P}(X, Z_j) = E_{i \ne j} \left[ {\rm{\ln\ }} P(X, Z) \right] + const$

Then:

\begin{align}
\mathcal{L}(Q) &= \int_{Z_j} Q_j(Z_j) {\rm {ln\ }} \tilde{P}(X, Z_j) dZ_j - \int_{Z_j} Q_j(Z_j) {\rm{\ ln}} Q_j(Z_j) dZ_j + const \\[0.5em]
&= -KL(Q_j(Z_j) \ \| \ \tilde{P}(X, Z_j))
\end{align}

And we can calculate all of these elements in this new KL divergence. Let the minimum be denoted $Q^{*}_j(Z_j)$. Then the log of the optimal solution is:

$${\rm {ln\ }} \tilde{P}(X, Z_j) = {\rm {ln\ }}(Q^{*}_j(Z_j)) = E_{i \ne j} \left[ {\rm{\ln\ }} P(X, Z) \right] + const$$


and so:
  

$${\rm {ln\ }}(Q^{*}_j(Z_j)) = E_{i \ne j} \left[ {\rm{\ln\ }} P(X, Z) \right] + const $$

$$Q^{*}_j(Z_j) = {\rm{exp}}\left\{ E_{i \ne j} \left[ {\rm{\ln\ }} P(X, Z) \right] + const \right\}$$

$$Q^{*}_j(Z_j)= {\rm{exp}}\left\{ E_{i \ne j} \left[ {\rm{\ln\ }} P(X, Z) \right]\right\}{\rm{exp}}\left\{const\right\}$$

    
and we choose ${\rm{exp}}\left\{const\right\}$ such that the distribution is normalized - that is, the final result is:
    
$$Q^{*}_j(Z_j)= \displaystyle{\frac{{\rm{exp}}\left\{ E_{i \ne j} \left[ {\rm{\ln\ }} P(X, Z) \right]\right\}}{\int_{Z_j} \rm{exp}\left\{ E_{i \ne j} \left[ {\rm{\ln\ }} P(X, Z) \right]\right\} d{Z_j}}}$$

So to arrive at a best distribution for all $Z_j,\ j = 1, \ldots n$, we need to initialize all of the factors $Q_i(Z_i), i \ne j$, then calculate the expectation of $Z_j$ with respect to the other factors. We then update our $Q_j(Z_j)$, and move to another distribution, holding all the others constant. In this way, we will arrive at an optimal distribution for $Q$, since the bound is convex with respect to each of the $Q_i(Z_i)$ distributions (Bishop, 2006).
